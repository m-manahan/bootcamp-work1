```python
#Weather API Homework Assignment
#by Marcus Manahan -  08/02/18

# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

size1 = 1500    #originally 1500
size2 = 1500    #originally 1500

#for faster testing

#size1 = 100    
#size2 = 100

from datetime import date
today = str(date.today())
print(today)   # '2017-12-26'

today_date = date.today().strftime('%d-%m-%Y')
```

```python
## Generate Cities List

# List for holding lat_lngs and cities
lat_lngs = []
cities = []
data_lat=[]
data_lng=[]
data_temp = []
data_humid = []
data_wind = []
data_cloud = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=size1)
lngs = np.random.uniform(low=-180.000, high=180.000, size=size2)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        data_lat.append(lat_lng[0])
        data_lng.append(lat_lng[1])

# Print the city count to confirm sufficient count
len(cities)
late = lat_lng[0]
leng = lat_lng[1]
cities[0]
cities = pd.DataFrame(cities)
cities
lats

response=[]
len(cities)
```

```python
cities["latitude"]=data_lat
cities["longitude"]=data_lng
cities.head()
cities = cities.rename(columns={0:"City"})
weather_df = cities
weather_df.head()
```

```python
weather_df["Temperature"] = ""
weather_df["Humidity"] = ""
weather_df["Wind Speed"] = ""
weather_df["Cloudiness"] = ""
```

```python
##Perform API Calls
for index, row in weather_df.iterrows():
# Starting URL for Weather Map API Call
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

    city = row["City"]    #pull name of city from the weather_df dataframe
    city = city.replace(" ","_")
    url = url + "&q=" + city                                
    #response = requests.get(url).json()
    response.append(requests.get(url).json())
    
    try:
        print(f"{index}. Getting weather information for the city of {city}.")
        print(f"url: {url}")
        #print(response[index]["main"]["temp"])
        data_temp.append(response[index]["main"]["temp"])
        data_humid.append(response[index]["main"]["humidity"])
        data_wind.append(response[index]["wind"]["speed"])
        data_cloud.append(response[index]["clouds"]["all"])
        
        #weather_df.loc[index, 'Temperature'] = response[index]["main"]["temperature"]
        #weather_df.loc[index, "Humidity"] = response[index]["main"]["humidity"] #humidity data
        #weather_df.loc[index, "Wind Speed"] = response[index]["wind"]["speed"] #wind speed
        #weather_df.loc[index, "Cloudiness"] = response[index]["clouds"]["all"] #
        
    except (KeyError, IndexError):                      #error handling - if the city is not in the OpenWeather API then, fill NA
        print("Missing field/result... skipping.")
        data_temp.append(np.nan)
        data_wind.append(np.nan)
        data_humid.append(np.nan)
        data_cloud.append(np.nan)
    print("------------")
print("Data Retrieval Complete!")
weather_df.head()
len(weather_df)
```
